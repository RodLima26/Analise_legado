# Script para associação das telas acessadas com a referência do site

```python
import pandas as pd
import numpy as np
```

```python
telas = pd.read_csv(r"C:\Users\rodli\OneDrive\Documentos\Mestrado\Arquivos\telas_acessadas.csv", sep=';', encoding='utf-8')
```

```python
referencia = pd.read_csv(r"C:\Users\rodli\OneDrive\Documentos\Mestrado\Arquivos\referencia.csv", sep=';', encoding='utf-8')
```

```python
var_link = telas['Página'].str[:4]
var_graf = telas['Página'].str[:7]
for ind in range(0, len(telas)):
    encontrou = 'N'
    for ind2 in range(0, len(referencia)):
        if telas['Código'][ind] == referencia['Codigo'][ind2]:
            if telas['Página'][ind] == referencia['Pagina'][ind2] or telas['Página'][ind] == referencia['Pagina2'][ind2]:
                referencia['Visualizacao'][ind2] = referencia['Visualizacao'][ind2] + telas['Visualizações'][ind]
                referencia['Usuario'][ind2] = referencia['Usuario'][ind2] + telas['Usuários'][ind]
                encontrou = 'S'
                
        if referencia['Pagina'][ind2] == telas['Página'][ind] and referencia['Codigo'][ind2] == 'NAO' and encontrou == 'N':
            referencia['Visualizacao'][ind2] = referencia['Visualizacao'][ind2] + telas['Visualizações'][ind]
            referencia['Usuario'][ind2] = referencia['Usuario'][ind2] + telas['Usuários'][ind]
            encontrou = 'S'
                
        if referencia['Pagina2'][ind2] == telas['Página'][ind] and referencia['Codigo'][ind2] == 'NAO' and encontrou == 'N':
            referencia['Visualizacao'][ind2] = referencia['Visualizacao'][ind2] + telas['Visualizações'][ind]
            referencia['Usuario'][ind2] = referencia['Usuario'][ind2] + telas['Usuários'][ind]
            encontrou = 'S'

        if var_link[ind] == 'link' and referencia['Pagina'][ind2] == telas['Página'][ind] and encontrou == 'N':
            referencia['Visualizacao'][ind2] = referencia['Visualizacao'][ind2] + telas['Visualizações'][ind]
            referencia['Usuario'][ind2] = referencia['Usuario'][ind2] + telas['Usuários'][ind]
            encontrou = 'S'

        if var_graf[ind] == 'grafico' and referencia['Pagina'][ind2] == telas['Página'][ind] and encontrou == 'N':
            referencia['Visualizacao'][ind2] = referencia['Visualizacao'][ind2] + telas['Visualizações'][ind]
            referencia['Usuario'][ind2] = referencia['Usuario'][ind2] + telas['Usuários'][ind]
            encontrou = 'S'

        if encontrou == 'N':
            if telas['Página'][ind] == referencia['Pagina'][ind2] or telas['Página'][ind] == referencia['Pagina2'][ind2]:
                referencia['Visualizacao'][ind2] = referencia['Visualizacao'][ind2] + telas['Visualizações'][ind]
                referencia['Usuario'][ind2] = referencia['Usuario'][ind2] + telas['Usuários'][ind]
                encontrou = 'S'

           
    if encontrou == 'N':
        print(telas['Página'][ind])
        print(telas['Visualizações'][ind])
        print(telas['Usuários'][ind])
                       
```
