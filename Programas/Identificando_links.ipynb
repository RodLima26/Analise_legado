# Importação de bibliotecas

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import os
```

# Identificar links com o comando url_for

```python
arquivos = os.listdir(r'C:\Users\rodli\PycharmProjects\fluzao\templates')
saida_final = []
for arq in range(1, len(arquivos)):
    arquivo = r'C:\\Users\\rodli\\PycharmProjects\\fluzao\\templates\\' + arquivos[arq]
    with open(arquivo, 'r', encoding="utf8") as fp:
        saida_final.append("Novo Arquivo")
        saida_final.append(arquivos[arq])
        soup = BeautifulSoup(fp)
        link = soup.find_all('a')
        array_saida = []
        saida2 = []
        for ind in range(0, len(link)):
            string = str(link[ind])
            for ind2 in range(0, len(string)):
                if string[ind2:ind2+7] == 'url_for':
                    ind3 = ind2 + 9
                    array_saida = []
                    while string[ind3] != "'":
                        array_saida.append(string[ind3])
                        ind3 = ind3 + 1
            link_site = ''.join(map(str, array_saida))
            saida2.append(link_site)
        saida2 = sorted(saida2)
        reg_ant = ''
        saida3 = []
        for dup in range(0, len(saida2)):
            if saida2[dup] != reg_ant:
                saida3.append(saida2[dup])
                reg_ant = saida2[dup]
        saida_final.append(saida3)

array_df = pd.DataFrame(saida_final, columns=['nome'])
```

# Salvando o resultado em arquivo csv

```python
ind = 0
lista_def = []
array_df = array_df.reset_index(drop=True)
while ind < len(array_df):
    ind_nome = ind + 1
    ind_link = ind + 2
    intermediario = (array_df['nome'][ind_nome], array_df['nome'][ind_link])
    ind = ind + 3
    lista_def.append(intermediario)

lista_final = pd.DataFrame(lista_def, columns=['HTML', 'Links'])
        
lista_final.to_csv(r'C:\Users\rodli\PycharmProjects\fluzao\static\graficos\lista_links.csv', sep=';')
```
